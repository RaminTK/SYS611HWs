```python
import numpy as np
from copy import deepcopy as dc
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
P95 = []
for run in range(10):
  rd = np.random.random(260)
  rs = np.random.random(260)
  t = list(range(261))
  D = []
  for ele in rd:
    if ele < 0.13:
      D.append(2)
    elif ele < 0.55:
      D.append(3)
    else:
      D.append(5)
  N = [40]
  td = [np.Infinity]
  P = [0]
  for i in range(260):
    temp = dc(N[-1])
    temp -= D[i]
    ttemp = dc(td[-1])
    if temp < 0:
      P.append(P[-1] - temp)
    else:
      P.append(P[-1])
    if temp < 30 and ttemp == np.Infinity:
      ttemp = t[i] + 5 + np.random.exponential(4)
    if ttemp <= t[i] + 1:
      temp += 40
      ttemp = np.Infinity
    N.append(temp)
    td.append(ttemp)
  print()
  print('simulation number:', run+1)
  plt.figure()
  plt.step(t, P)
  plt.xlabel('Time in Days')
  plt.ylabel('Penalty')
  plt.show()

  plt.figure()
  plt.step(t, N)
  plt.xlabel('Time in Days')
  plt.ylabel('Inventory Level (N)')
  plt.show()
  P95.append(P[-1])
print(P95)
print('95% confidence interval for the total penalty P after a 52-week (260 day) simulation is in range of :',np.mean(P95)- 1.96 * stats.sem(P95), np.mean(P95)+1.96*stats.sem(P95))
```

The results of simulations presented in the graphs show that the level of supply copmared to the demand is not eanough is most of the time of the period. It results to negative level of inventory leading to increasing amount of penalty. We should increase the supply level to solve the problem.

```python
P95 = []
for run in range(10):
  rd = np.random.random(260)
  rs = np.random.random(260)
  t = list(range(261))
  D = []
  for ele in rd:
    if ele < 0.13:
      D.append(2)
    elif ele < 0.55:
      D.append(3)
    else:
      D.append(5)
  N = [40]
  td = [np.Infinity]
  P = [0]
  for i in range(260):
    temp = dc(N[-1])
    temp -= D[i]
    ttemp = dc(td[-1])
    if temp < 0:
      P.append(P[-1] - temp)
    else:
      P.append(P[-1])
    if temp < 30 and ttemp == np.Infinity:
      ttemp = t[i] + 5 + np.random.exponential(4)
    if ttemp <= t[i] + 1:
      temp += 70
      ttemp = np.Infinity
    N.append(temp)
    td.append(ttemp)
  print()
  print('simulation number:', run+1)
  plt.figure()
  plt.step(t, P)
  plt.xlabel('Time in Days')
  plt.ylabel('Penalty')
  plt.show()

  plt.figure()
  plt.step(t, N)
  plt.xlabel('Time in Days')
  plt.ylabel('Inventory Level (N)')
  plt.show()
  P95.append(P[-1])
print('95% confidence interval for the total penalty P after a 52-week (260 day) simulation is in range of :',np.mean(P95)- 1.96 * stats.sem(P95), np.mean(P95)+1.96*stats.sem(P95))
```
