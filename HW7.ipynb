i) Considering +- two percent error of measurement I generated 1000 random volumes based on a triangular distribution.

```python
rand1 = np.random.random(1000)
c = 946
a = c * 0.98
b = c * 1.02
V = []
for r in rand1:
  if r < 2/(b-a):
    V.append(a + np.sqrt(r*(b-a)*(c-a)))
  else:
    V.append(b - np.sqrt((1-r)*(b-a)*(b-c)))
```

ii) For the volume of candy, I consider the diameter of the base of the cone 2.2 mm and the height if it 2.35 based on my measurment. Volume for the cone is $\frac { 1 } { 3 } \pi r ^{ 2 } h$. For this measurments I considered +- 10 percent error as well. Them I generated 1000 random r and h based on a triangular distribution. Then I calculated the volume using each couple of r and h.

```python
hc = 2.35
ha = hc * 0.9
hb = hc * 1.1
d = 2.2
rc = d/2
print (rc)
ra = rc * 0.9
rb = rc * 1.1

rand2 = np.random.random(1000)

h = []
for rnd in rand2:
  if rnd < 2/(hb-ha):
    h.append(ha + np.sqrt(rnd*(hb-ha)*(hc-ha)))
  else:
    h.append(hb - np.sqrt((1-rnd)*(hb-ha)*(hb-hc)))

rand3 = np.random.random(1000)

r = []
for rnd in rand3:
  if rnd < 2/(rb-ra):
    r.append(ra + np.sqrt(rnd*(rb-ra)*(rc-ra)))
  else:
    r.append(rb - np.sqrt((1-rnd)*(rb-ra)*(rb-rc)))

V_cone = []
for i in range(1000):
  V_cone.append((1/3)*np.pi*r[i]*r[i]*h[i])
np.array(V_cone)
```

iii) I used the same method and error to calculate mu.

```python
mc = 0.55
ma = mc * 0.9
mb = mc * 1.1

rand4 = np.random.random(1000)

m = []
for rnd in rand4:
  if rnd < 2/(mb-ma):
    m.append(ma + np.sqrt(rnd*(mb-ma)*(mc-ma)))
  else:
    m.append(mb - np.sqrt((1-rnd)*(mb-ma)*(mb-mc)))
np.array(m)
```

```python
from scipy.optimize import minimize

def obj(c):
    """objective function, to be solved."""
    sol = np.prod([2*(d-c)/((b-c)*(b-a)) if (d >= c and d <= b) else 2*(c-d)/((c-a)*(b-a)) if (d >= a and d < c) else 0 for d in eruptions])
    return  - np.log(sol)
solution = (minimize(obj, [(a+b)/2]))

logL = - solution.fun
c = solution.x

print('a = ',a,'b = ', b,'c = ', c[0],' log L = ', logL )
```
