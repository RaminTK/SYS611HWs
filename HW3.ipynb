# 3.1 Simulating Ï€ Again 

**Part a**

```python
X = np.random.random(1000)
Y = np.random.random(1000)
Z = []
for i in range(len(X)):
  if np.sqrt(X[i]**2 + Y[i]**2) <= 1 :
    Z.append(4)
  else:
    Z.append(0)
print('list of 10 first X:',X[0:10])
print()
print('list of 10 first Y:',Y[0:10])
print()
print('list of 10 first Z:',Z[0:10])
print(len(Z))
```

Part b

```python

confidence_level = 0.05
z_crit = st.norm.ppf(1-confidence_level/2)
E_Z = np.array([np.average(Z[0:i]) for i in range(len(Z))])
Z_ci = z_crit*np.array([st.sem(Z[0:i]) for i in range(len(Z))])

plt.figure(figsize = (14, 8))
plt.plot(range(len(Z)), E_Z, 
         'b', label='Mean Estimate')
plt.plot(range(len(Z)), E_Z - Z_ci, 
         'g', label='95% CI Lower Bound')
plt.plot(range(len(Z)), E_Z + Z_ci, 
         'r', label='95% CI Upper Bound')
plt.xlabel('Z')
plt.ylabel('Estimate of $\pi$')
plt.legend(loc='best')

print((E_Z - Z_ci)[-1], E_Z[-1], (E_Z + Z_ci)[-1])
```

The CDF for D:

\begin{cases}
    0 & \text{for } D < 40, \\[2pt]
    \frac{(D-40)^2}{(90-40)(70-40)} & \text{for } 40 \leq D \leq 70, \\[4pt]
    1-\frac{(90-d)^2}{(90-40)(90-70)} & \text{for } 70 < D \leq 90, \\[4pt]
    1 & \text{for } 90 < D.
\end{cases}
              


Given a random variate r drawn from the uniform distribution in the interval (0, 1), then the variate

\begin{cases}
    a+{\sqrt {r(b-a)(c-a)}}&{\text{ for }}0<r<F(c)\\&\\b-{\sqrt {(1-r)(b-a)(b-c)}}&{\text{ for }}F(c)\leq r<1
\end{cases}


where ${\displaystyle F(c)=(c-a)/(b-a).}$

In this question, $a = 40, b = 90, and c = 70$. Then:

${\displaystyle F(70)=(70-40)/(90-40) = 0.6.}$

```python
def demand(rand):
  if rand <= ((70-40)/(90-40)):
    D = 40 + np.round(np.sqrt(rand * (90-40) *(70 -40)))
  else:
    D = 90 - np.round(np.sqrt((1-rand) *(90 - 40) * (90-70)))
  return D
```

```python
def profit(O,D):
  if D >= O:
    P = (12-10) * O
  else:
    P = 12*D + 3 * (O - D) - 10 *O
  return P
```
